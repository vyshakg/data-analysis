```python
import pandas as pd
import numpy as np
```

## Reading csv and filtering value as required

```python
df = pd.read_csv("exp.csv",names=['edate','statement','dr','cr','bal'],header=0)
df["edate"] = pd.to_datetime(df["edate"],format="%d-%m-%Y")
df[["dr", "cr"]] = df[["dr", "cr"]].replace(" ",0).apply(pd.to_numeric)
```

```python
df =  df[df['edate'].dt.year==2018]
df.tail(10)
```

```python
exp_df = df[['edate','dr','cr']]

exp_df = exp_df.groupby(exp_df['edate'].dt.month,as_index=False).agg(['sum'])
exp_df.columns = exp_df.columns.map('_'.join)
exp_df.insert(0,'month',exp_df.index)
# exp_df.to_json(orient='records')
```

```python
import re

def filter(str):
    strList = " ".join(str.split("/")).split()
    copy_List = []
    for l in strList:
        if not bool(re.search(r'\d', l)):
            copy_l =re.sub(r"\s+","_",l.upper())
            copy_List.append(copy_l)  
            
    return " ".join(copy_List)
```

```python
df_month =  df[df['edate'].dt.month==5]
# df_month.tail(10)
df_month['statement'] = df_month['statement'].apply(filter)
df_month['statement'].replace('', np.nan, inplace=True)
df_month.dropna(inplace=True)
df_month.tail(10)
# pd.Series(' '.join(df_month['statement']).lower().split("/")).value_counts()[:100]
```

```python
df_month['statement'].value_counts()
```

```python
cat_df = df_month[['edate','statement','dr','cr']]

cat_df = cat_df.groupby(cat_df['statement'],as_index=False).agg(['sum','count'])
cat_df.columns = cat_df.columns.map('_'.join)
cat_df.insert(0,'particulars',cat_df.index)
cat_df.sort_values('cr_count',ascending=False, inplace=True)
cat_df = cat_df.drop(['dr_count'],axis="columns")
cat_df = cat_df.reset_index(drop=True)

cat_df[:25]
```

```python
Tran = ['POS','UPI','ATM-CASH','NEFT','ECOM','MOB']
cat_df['categories'] = cat_df['particulars']
def cat_filter(str):
    for word in Tran:
        if word in str:
            return word
    return 'OTHERS'


```

```python
cat_df['categories'] = cat_df['categories'].apply(cat_filter)
cat_df
```
