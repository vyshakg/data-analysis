```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("city.csv")
```

```python
df =df[["name_of_city","state_code","state_name","location"]]
```

## STATE SQL

```python
df_state= df[["state_code","state_name"]]
df_state = df_state.groupby(["state_code","state_name"]).count()
df_state.reset_index(inplace=True)
df_state.rename(columns={'state_code':'id'},inplace=True)
df_state.set_index('id',inplace=True)
```

```python
def frame_to_sql(source):
    f = open('state.sql', 'w')
    f.write('INSERT INTO state ('+ 'id' + str(', '.join(source.columns)) + ') VALUES ')
    f.write('\n')
    count=1
    for index, row in source.iterrows(): 
        lst = np.insert(row.values, 0, index, axis=0)
        f.write( str(tuple(lst)))
        f.write(';') if count==len(source) else f.write(',\n')
        count=count+1
        
    f.close()
```

```python
frame_to_sql(df_state)
```

## CITY SQL

```python
df_city = df[["name_of_city","state_code","location"]]
df_city[["latitude","longitude"]] = df['location'].str.split(',', 1, expand=True)
df_city.drop('location', axis=1, inplace=True)
df_city.rename(columns={'name_of_city':'city_name','state_code':'state_id'},inplace=True)
df_city = df_city[['city_name','latitude','longitude','state_id']]
df_city.index+= 1 
# df_city.set_index('id',inplace=True)
```

```python
def frame_to_sql(source):
    f = open('city.sql', 'w')
    f.write('INSERT INTO city ('+ 'id,' + str(', '.join(source.columns)) + ') VALUES ')
    f.write('\n')
    count=1
    for index, row in source.iterrows(): 
        lst = np.insert(row.values, 0, index, axis=0)
        f.write( str(tuple(lst)))
        f.write(';') if count==len(source) else f.write(',\n')
        count=count+1
        
    f.close()
```

```python
frame_to_sql(df_city)
```
