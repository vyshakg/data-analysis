```python
def filter(str):
    lst = str.split(",")
    if len(lst) == 3:
        state=lst[1]
    else:
        state = lst[0]
    return state
```

```python
df_city["faddress"] = df_city["faddress"].apply(filter)
```

```python
import re
def filter2(str):
    lst = str.split(" ")
    for l in lst:
        if bool(re.search(r'\d', l)):
            return ' '.join(lst[:-1])
    return str
```

```python
df_city.loc[:,"state_name"] = df_city["faddress"].apply(filter2)
df_city.head(2)
```
